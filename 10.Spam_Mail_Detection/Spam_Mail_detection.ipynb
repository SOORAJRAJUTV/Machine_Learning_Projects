```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from  sklearn.metrics import accuracy_score
```

```python
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
raw_mail_data.head()
```

```python
raw_mail_data.shape
```

```python
raw_mail_data['Category'].value_counts()
```

```python
raw_mail_data.loc[raw_mail_data['Category'] == 'ham','Category'] = 0
raw_mail_data.loc[raw_mail_data['Category'] == 'spam','Category'] = 1
```

```python
raw_mail_data.head()
```

**ham --> 0**

**spam -->1**

```python
X = raw_mail_data['Message']
Y = raw_mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=2, test_size=0.2,stratify=Y)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df=1,stop_words = 'english',lowercase = True)
```

```python
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features  = feature_extraction.transform(X_test)
```

```python
print(X_train_features)
```

```python
Y_train = Y_train.astype('int')
Y_test =  Y_test.astype('int')
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

```python
training_data_prediction = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train,training_data_prediction)
print('Accuracy score on training data',round(accuracy_on_training_data*100,2))
```

```python
test_data_prediction = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test,test_data_prediction)
print('Accuracy score on test data',round(accuracy_on_test_data*100,2))
```
