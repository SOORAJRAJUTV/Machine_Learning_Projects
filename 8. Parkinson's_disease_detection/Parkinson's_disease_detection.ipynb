```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('/content/parkinsons.data')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset['status'].value_counts()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
X = dataset.drop(['name','status'], axis=1)
Y = dataset['status']
```

```python
print(X)
```

```python
print(X.std())
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

```python
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
print(X)
```

```python
X.std()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2, stratify = Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = SVC(kernel='linear')
model.fit(X_train, Y_train)
training_data_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,training_data_prediction)
print(training_data_accuracy*100)
```

```python
test_data_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,test_data_prediction)
print(test_data_accuracy*100)
```
